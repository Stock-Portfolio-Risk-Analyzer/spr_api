```python
from datautils import quandl_info as gd
from datetime import datetime as dt
from IPython.display import display
import pandas as pd
import numpy as np
```

## Get daily stock data between dates

```python
start_date = dt(year=2015, month=1, day=1)
end_date = dt(year=2015, month=12, day=31)

data = gd.get_stock_data('GOOG', start_date, end_date)
display(data.head(10))
```

## Get multiple stock data

```python
start_date = dt(year=2015, month=1, day=1)
end_date = dt(year=2015, month=12, day=31)

data = gd.get_stock_data_multiple(["GOOG", "AAPL"], start_date, end_date)
display(data["GOOG"].head(10))
display(data["AAPL"].head(10))
```

## Get percentage returns

```python
start_date = dt(year=2015, month=1, day=1)
end_date = dt(year=2015, month=12, day=31)

data = gd.get_pct_returns("GE", start_date, end_date)
display(data.head(10))
```

## Use a different Quandl database

```python
start_date = dt(year=2015, month=1, day=1)
end_date = dt(year=2015, month=12, day=31)

# use Yahoo's stock database
# default is Quandl's WIKI
data = gd.get_stock_data("GOOG", start_date, end_date, "YAHOO")
display(data.head(10))
```
