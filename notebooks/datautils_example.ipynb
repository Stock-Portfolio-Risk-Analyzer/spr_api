# Get daily stock data between dates

```python
start_date = dt(year=2015, month=1, day=1)
end_date = dt(year=2015, month=12, day=31)

data = yf.get_stock_data('GOOG', start_date, end_date)
display(data.head(10))
```

```python
pct_returns = yf.get_pct_returns('GOOG', start_date, end_date, col='Adj Close')
data['pct_returns'] = pct_returns

returns = yf.get_returns('GOOG', start_date, end_date, col='Adj Close')
data['returns'] = returns

log_returns = np.log(data['Adj Close']) - np.log(data['Adj Close'].shift(1)).fillna(0)
data['log_returns'] = log_returns

display(data.head(10))
```
